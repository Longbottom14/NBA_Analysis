```python
import requests
import os
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
url_teams = "https://www.basketball-reference.com/teams/"
```

```python
data_teams = requests.get(url_teams)
```

```python
fault=[]
year="2022"
for link1,link2 in zip(fault_links_teams1,fault_links_teams2):
    #make directories
    parent_dir="C:/Users/user/Solo Dolo/NBA_Analysis/Game"
    team_name = link1.split('/')[-2]#"ATL"
    folder_path = os.path.join(parent_dir,team_name)
    os.mkdir(folder_path)
    print(f"made new directory at : {folder_path}")
    print()
    
    #try and error
    def try_and_error(func):
        try:
            func(link=link1)
        except ValueError:
            try:
                func(link=link2)
            except ValueError:
                print(f'Problem with {team_name}')
                fault.append(team_name)
                #continue #
    #gamesdata
    def games(link):
        standings_url_games = "https://www.basketball-reference.com"+link+year+"_games.html"
        data_games = requests.get(standings_url_games)
        time.sleep(1)
        matches_games = pd.read_html(data_games.text, match="Regular Season")[0]
        print("games scraped succesfully")
        print(f'shape of {team_name} data :{matches_games.shape}')
        games_path_csv = os.path.join(folder_path,'games.csv')
        matches_games.to_csv(games_path_csv)
        print(f"Saved csv as  : {games_path_csv}")
        print()
    try_and_error(func=games)

    def gamelog(link):
        standings_url_gamelog = "https://www.basketball-reference.com"+link+year+"/gamelog/"
        data_gamelog = requests.get(standings_url_gamelog)
        time.sleep(1)
        matches_gamelog_stats = pd.read_html(data_gamelog.text, match="Regular Season")[0]
        print("gamelog_stats scraped succesfully")
        print(f'shape of {team_name} data :{matches_gamelog_stats.shape}')
        gamelog_path_csv = os.path.join(folder_path,'gamelog.csv')
        matches_gamelog_stats.to_csv(gamelog_path_csv)
        print(f"Saved csv as  : {gamelog_path_csv}")
        print()
    try_and_error(func=gamelog)
    
    def gamelog_advanced(link):
        standings_url_advanced = "https://www.basketball-reference.com"+link+year+"/gamelog-advanced/"
        data_advanced = requests.get(standings_url_advanced)
        time.sleep(1)
        matches_advance_stats = pd.read_html(data_advanced.text, match="Regular Season")[0]
        print("games_advance_stats scraped succesfully")
        print(f'shape of {team_name} data :{matches_advance_stats.shape}')
        game_advance_path_csv = os.path.join(folder_path,'game_advance_stats.csv')
        matches_advance_stats.to_csv(game_advance_path_csv)
        print(f"Saved csv as  : {gamelog_path_csv}")
        print()
    try_and_error(func=gamelog_advanced)
    
    def starting_lineup(link):
        standings_url_starting_lineup = "https://www.basketball-reference.com"+link+year+"_start.html"
        data_starting_lineup = requests.get(standings_url_starting_lineup)
        time.sleep(1)
        matches_starting_lineup = pd.read_html(data_starting_lineup.text, match="Starting Lineups")[0]
        print("matches_starting_lineup scraped succesfully")
        print(f'shape of {team_name} data :{matches_starting_lineup.shape}')
        starting_lineup_path_csv = os.path.join(folder_path,'games_starting_lineup.csv')
        matches_starting_lineup.to_csv(starting_lineup_path_csv)
        print(f"Saved csv as  : {starting_lineup_path_csv}")
        print()
    try_and_error(func=starting_lineup)
    
    def prev_season_func(link):
        prev_season_url = "https://www.basketball-reference.com"+link
        prev_season_data = requests.get(prev_season_url)
        time.sleep(1)
        prev_season = pd.read_html(prev_season_data.text, match="Seasons")[0]
        print("matches_prev_season scraped succesfully")
        print(f'shape of {team_name} data :{prev_season.shape}')
        prev_season_path_csv = os.path.join(folder_path,'prev_season.csv')
        prev_season.to_csv(prev_season_path_csv)
        print(f"Saved csv as  : {prev_season_path_csv}")
        print()
    try_and_error(func=prev_season_func)
    
    def H2H_func(link):
        H2H_url ="https://www.basketball-reference.com"+link+"head2head.html"
        H2H_data = requests.get(H2H_url)
        time.sleep(1)
        H2H = pd.read_html(H2H_data.text, match="Opponents")[0]
        print("matches_H2H scraped succesfully")
        print(f'shape of {team_name} data :{H2H.shape}')
        H2H_path_csv = os.path.join(folder_path,'H2H.csv')
        H2H.to_csv(H2H_path_csv)
        print(f"Saved csv as  : {H2H_path_csv}")
        print()
    try_and_error(func=H2H_func)
    
    def coaches_func(link):
        coaches_url = "https://www.basketball-reference.com"+link+"coaches.html"
        coaches_data = requests.get(coaches_url)
        time.sleep(1)
        coaches = pd.read_html(coaches_data.text, match="Coaches")[0]
        print("team_coaches scraped succesfully")
        print(f'shape of {team_name} data :{coaches.shape}')
        coaches_path_csv = os.path.join(folder_path,'coaches.csv')
        coaches.to_csv(coaches_path_csv)
        print(f"Saved csv as  : {coaches_path_csv}")
        print()
    try_and_error(func=coaches_func)

    def referees_func(link):
        referees_url = "https://www.basketball-reference.com"+link+year+"_referees.html"
        referees_data = requests.get(referees_url)
        time.sleep(1)
        referees = pd.read_html(referees_data.text, match="Referees")[0]
        print("matches_referees scraped succesfully")
        print(f'shape of {team_name} data :{referees.shape}')
        referees_path_csv = os.path.join(folder_path,'referees.csv')
        referees.to_csv(referees_path_csv)
        print(f"Saved csv as  : {referees_path_csv}")
        print()
    try_and_error(func=referees_func)
    
```
