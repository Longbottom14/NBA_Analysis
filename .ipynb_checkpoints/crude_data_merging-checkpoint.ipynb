```python
def clean_H2H(H2H_df):#,games_df):
    H2H_df= H2H_df.copy()
    print("before")
    display(H2H_df[H2H_df.Franchise=='Denver Nuggets'])
    #games_df = games_df.copy()
    #h1 = H2H_df[H2H_df.Franchise.isin(games_df.Opponent.unique())].reset_index(drop=True)
    #repeated = h1.Franchise.value_counts()[h1.Franchise.value_counts()>1].index[0]
    val_count = H2H_df.Franchise.value_counts()
    if len(val_count[H2H_df.Franchise.value_counts()>1])>0:
        max_game = H2H_df[H2H_df.Franchise=='Denver Nuggets']['G'].max()
        H2H_df = H2H_df.drop(H2H_df[(H2H_df.Franchise=='Denver Nuggets')&(H2H_df.G!=max_game)].index)
    print()
    print("after")
    display(H2H_df[H2H_df.Franchise=='Denver Nuggets'])
    return H2H_df    
```
